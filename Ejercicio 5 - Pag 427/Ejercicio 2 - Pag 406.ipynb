```python
import tkinter as tk  # Importamos la librería para hacer la interfaz gráfica


class Vendedor:
    def __init__(self, nombre, apellidos):
        
        self.nombre = nombre
        self.apellidos = apellidos
        self.edad = None  

    # Este método sirve para verificar si la edad es válida
    def verificar_edad(self, edad):
        # Si la edad es menor que 0 o mayor que 120, no es válida
        if edad < 0 or edad > 120:
            raise ValueError("La edad no puede ser negativa ni mayor a 120.")
        
        # Si la edad es 18 o menos, no puede trabajar como vendedor
        if edad <= 18:
            raise ValueError("El vendedor debe ser mayor de 18 años.")

        # Si pasa las validaciones, se guarda la edad en el objeto
        self.edad = edad

    # Este método devuelve el texto con la información del vendedor
    def imprimir(self):
        resultado = " - Nombre: " + self.nombre + "\n"
        resultado += " - Apellidos: " + self.apellidos + "\n"
        resultado += " - Edad: " + str(self.edad)
        return resultado

# Aquí empieza el código de la interfaz
def crear_interfaz():
    
    def al_hacer_click():
        
        texto_nombre = campo_nombre.get()
        texto_apellidos = campo_apellidos.get()
        texto_edad = campo_edad.get()

        try:
            
            edad_convertida = int(texto_edad)

            
            vendedor = Vendedor(texto_nombre, texto_apellidos)

           
            vendedor.verificar_edad(edad_convertida)

            
            mensaje = vendedor.imprimir()
            mostrar_resultado_en_otra_ventana(mensaje, True)

        except ValueError as error:
            #
            mensaje_error = " X " + str(error)
            mostrar_resultado_en_otra_ventana(mensaje_error, False)

    
    def mostrar_resultado_en_otra_ventana(texto, es_valido):
        nueva_ventana = tk.Toplevel(ventana_principal)
        nueva_ventana.title("Resultado")

        
        if es_valido:
            color_fondo = "#f0fff0"  
            color_texto = "black"
        else:
            color_fondo = "#ffecec"  
            color_texto = "red"


        etiqueta_titulo = tk.Label(nueva_ventana, text="Resultado", font=("Helvetica", 14, "bold"))
        etiqueta_titulo.pack(pady=10)

        
        etiqueta_resultado = tk.Label(nueva_ventana, text=texto, bg=color_fondo, fg=color_texto,
                                      font=("Courier", 11), width=42, height=6, wraplength=320,
                                      justify="left", bd=2, relief="sunken", anchor="nw")
        etiqueta_resultado.pack(padx=15, pady=10)

    # Ahora creamos la ventana principal
    ventana_principal = tk.Tk()
    ventana_principal.title("Formulario del Vendedor")
    ventana_principal.geometry("400x400")
    ventana_principal.configure(bg="#eaf6f6")  # Color suave de fondo

    
    etiqueta_titulo = tk.Label(ventana_principal, text="Formulario del Vendedor",
                               font=("Helvetica", 16, "bold"), bg="#eaf6f6")
    etiqueta_titulo.pack(pady=10)

  
    marco_formulario = tk.Frame(ventana_principal, bg="#eaf6f6")
    marco_formulario.pack()

    etiqueta_nombre = tk.Label(marco_formulario, text="Nombre:", font=("Helvetica", 11), bg="#eaf6f6")
    etiqueta_nombre.pack(anchor="w", padx=20)
    campo_nombre = tk.Entry(marco_formulario, font=("Helvetica", 11), width=30)
    campo_nombre.pack(padx=20, pady=5)

    
    etiqueta_apellidos = tk.Label(marco_formulario, text="Apellidos:", font=("Helvetica", 11), bg="#eaf6f6")
    etiqueta_apellidos.pack(anchor="w", padx=20)
    campo_apellidos = tk.Entry(marco_formulario, font=("Helvetica", 11), width=30)
    campo_apellidos.pack(padx=20, pady=5)

   
    etiqueta_edad = tk.Label(marco_formulario, text="Edad:", font=("Helvetica", 11), bg="#eaf6f6")
    etiqueta_edad.pack(anchor="w", padx=20)
    campo_edad = tk.Entry(marco_formulario, font=("Helvetica", 11), width=30)
    campo_edad.pack(padx=20, pady=5)

    # Botón para validar los datos
    boton_validar = tk.Button(ventana_principal, text="Validar datos", command=al_hacer_click,
                              bg="#4CAF50", fg="white", font=("Helvetica", 11, "bold"),
                              relief="flat", padx=10, pady=5)
    boton_validar.pack(pady=15)

    
    ventana_principal.mainloop()

crear_interfaz()
```
