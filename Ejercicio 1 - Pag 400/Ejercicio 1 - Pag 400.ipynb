```python
import tkinter as tk

# Función principal: se encarga de leer datos, calcular y mostrar el resultado

def calcular_paso_a_paso():
    
    num_texto = entry_num.get()
    den_texto = entry_den.get()
    
    try:
        numerador_valor = float(num_texto)
    except ValueError:
        # Si falla la conversión del numerador:
        entry_res.delete(0, tk.END)
        entry_res.insert(0, "Error: numerador no numérico")
        print("Error al convertir numerador:", num_texto)
        return  
    
    try:
        denominador_valor = float(den_texto)
    except ValueError:
        # Si falla la conversión del denominador:
        entry_res.delete(0, tk.END)
        entry_res.insert(0, "Error: denominador no numérico")
        print("Error al convertir denominador:", den_texto)
        return  
    
    try:
        division_resultado = numerador_valor / denominador_valor
    except ZeroDivisionError:
        
        division_resultado = None
        infinity_texto = "Infinity"
        entry_res.delete(0, tk.END)
        entry_res.insert(0, infinity_texto)
        print("División por cero detectada. Mostrando Infinity.")
    else:
        
        texto_resultado = str(division_resultado)
        entry_res.delete(0, tk.END)
        entry_res.insert(0, texto_resultado)
        print("División exitosa:", numerador_valor, "/", denominador_valor, "=", division_resultado)
    finally:
        # Esto siempre se ejecuta, haya excepción o no
        print(">> Bloque finally ejecutado, fin de calcular_paso_a_paso()")


# Función para borrar todo y dejar limpio

def borrar_campos():
    entry_num.delete(0, tk.END)
    entry_den.delete(0, tk.END)
    entry_res.delete(0, tk.END)
    print("Se limpiaron los campos de la interfaz")

# Configuración de la ventana principal 

root = tk.Tk()
root.title("Ejercicio")
root.resizable(False, False)
root.geometry("300x180")  # fijo ancho x alto


label1 = tk.Label(root, text="Numerador:")
label1.place(x=20, y=20)
entry_num = tk.Entry(root)
entry_num.place(x=110, y=20, width=150)


label2 = tk.Label(root, text="Denominador:")
label2.place(x=20, y=55)
entry_den = tk.Entry(root)
entry_den.place(x=110, y=55, width=150)


label3 = tk.Label(root, text="Resultado:")
label3.place(x=20, y=90)
entry_res = tk.Entry(root)
entry_res.place(x=110, y=90, width=150)


btn_calcular = tk.Button(root, 
                         text="Calcular", 
                         command=calcular_paso_a_paso)
btn_calcular.place(x=60, y=130, width=80)


btn_borrar = tk.Button(root, 
                       text="Borrar", 
                       command=borrar_campos)
btn_borrar.place(x=160, y=130, width=80)

root.mainloop()
```
